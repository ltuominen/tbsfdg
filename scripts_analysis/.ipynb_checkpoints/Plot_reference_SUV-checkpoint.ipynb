```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel
```

```python
data = pd.read_csv('wmgm_mean_activity_values.txt', delimiter=' ', header=None)
data.columns =['subject', 'session', 'value']
data['subject_session'] = data.subject + data.session.str.replace('ses', '')
data.drop(columns=['subject', 'session'], inplace=True)
```

```python
# separate active and sham

unmask = pd.read_csv('../info/unmask.txt', delimiter=' ', dtype='str')
active = unmask.subject + '-' + unmask.active_session
sham = unmask.subject + '-' + unmask.sham_session
data_active = data[data.subject_session.isin(active)]
data_sham = data[data.subject_session.isin(sham)]
```

```python
df = pd.DataFrame({'subject': unmask.subject, 
                   'SUV_active': data_active.value.to_list(), 
                   'SUV_sham': data_sham.value.to_list() })
df = df.melt(id_vars='subject', value_vars=['SUV_active', 'SUV_sham'], 
                            var_name='condition', value_name='SUV')
```

```python
print(ttest_rel(data_active.value,data_sham.value))
```

```python
plt.figure(figsize=(5, 3))
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False

# plot
palette = ['salmon', 'deepskyblue']
p=sns.boxplot(data=df, x='condition', y='SUV', palette=palette, width= 0.3, fliersize=0)
sns.swarmplot(data=df, x='condition', y='SUV', size=8, linewidth=1,edgecolor='black', color='white');
```
