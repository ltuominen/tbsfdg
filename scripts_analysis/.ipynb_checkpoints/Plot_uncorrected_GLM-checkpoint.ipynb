```python
from nilearn.plotting import plot_stat_map
import os
import numpy as np
import nibabel as nb
import matplotlib.pyplot as plt
from nibabel.affines import apply_affine
```

```python
fsl_dir=os.environ['FSL_DIR']
mni152_brain=fsl_dir + '/data/standard/MNI152_T1_2mm_brain.nii.gz'
```

```python
mni152_mask=nb.load(fsl_dir + '/data/standard/MNI152_T1_2mm_brain_mask.nii.gz').get_fdata()

data = nb.load('../group/active_vs_sham/osgm/sig.nii.gz').get_fdata()
affine = nb.load('../group/active_vs_sham/osgm/sig.nii.gz').affine
masked = data * mni152_mask
img = nb.Nifti1Image(masked, affine)
```

## Plot uncorrected GLM

```python
#fig, axes = plt.subplots(nrows=2, ncols=1)
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 5))

display = plot_stat_map(
    img,
    bg_img=mni152_brain,
    threshold=1.3,
    draw_cross=False,
    alpha = 0.6,
    display_mode='z',
    cut_coords=(-18, -15, -12, 15, 20),
    axes=axes[0]
);

display = plot_stat_map(
    img,
    bg_img=mni152_brain,
    threshold=1.3,
    draw_cross=False,
    alpha = 0.6,
    display_mode='x',
    cut_coords=(-5, -2, 0, 2, 5),
    axes=axes[1]
);
```

## Find min peak

```python
xyz = np.concatenate(np.where(masked == np.min(masked)))
coord = apply_affine(affine,xyz)
```

```python
# plot peak
display = plot_stat_map(
    img,
    bg_img=mni152_brain,
    threshold=1.3,
    draw_cross=True,
    alpha = 0.6,
    cut_coords=(coord),
);
```
