```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel
```

```python
data = pd.read_csv('../qc/fd_pet_motion.csv')
data.drop(columns="Unnamed: 0", inplace=True)
unmask = pd.read_csv('../info/unmask.txt', delimiter=' ', dtype='str')
```

```python
#mean=data.mean(axis=0)
#max=data.max(axis=0)
#mt2 = data > 2
#mt2 = mt2.sum(axis=0)
```

```python
# separate active and sham 
active = unmask.subject + '-' + unmask.active_session
sham = unmask.subject + '-' + unmask.sham_session
data_active = data[active]
data_sham = data[sham]
```

```python
# into df
dict_plot = {'subject':  unmask.subject, 'mean_active': data_active.mean(axis=0).to_list(), 'mean_sham': data_sham.mean(axis=0).to_list(),
             'max_active': data_active.max(axis=0).to_list(), 'max_sham': data_sham.max(axis=0).to_list()}
df=pd.DataFrame(dict_plot)

df = df.melt(id_vars='subject', value_vars=['mean_active', 'mean_sham', 'max_active', 'max_sham'], 
                            var_name='condition', value_name='value')
```

## Paired t-test differences in means and maxes 

```python
_,p_mean = ttest_rel(data_active.mean(axis=0),data_sham.mean(axis=0))
_,p_max = ttest_rel(data_active.max(axis=0),data_sham.max(axis=0))

print(p_mean)
print(p_max)
```

## Plot means and maxes 

```python
plt.figure(figsize=(10, 6))
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False

# plot
palette = ['salmon', 'deepskyblue', 'salmon', 'deepskyblue']
p=sns.boxplot(data=df, x='condition', y='value', palette=palette, width= 0.3, fliersize=0)
sns.swarmplot(data=df, x='condition', y='value', size=8, linewidth=1,edgecolor='black', color='white')

# decorations 
plt.xlabel('Condition', fontsize=15)  
p.set_xticks(range(4), labels=['active (mean)', 'sham (mean)', 'active (max)', 'sham (max)'], fontsize=12)
p.spines['bottom'].set_bounds(-0.25, 3.25)
plt.ylabel('Framewise displacement (FD)', fontsize=15)  
plt.title('Framewise motion during PET scans (15 2min frames)', fontsize=18) 

# annotate max values 
max_max_sham = df[df.condition == 'max_sham'].value.max()
max_max_active = df[df.condition == 'max_active'].value.max()

subj_max_sham = df[df.value==max_max_sham].subject.to_string(index=False)
subj_max_active = df.loc[df.value==max_max_active,'subject'].to_string(index=False)

p.text(2.1, max_max_active, subj_max_active )
p.text(3.1, max_max_sham, subj_max_sham )

plt.show();
```
